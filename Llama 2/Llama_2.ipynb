```python
task_template = """
Below is an instruction that describes a sequence classification task...

### Instruction:
Analyze the following input and classify them as World or Sports or Business or Sci/Tech.

### Input:
{input}

### Response:
{response}
"""
```

```python
from langchain.prompts import ChatPromptTemplate, PromptTemplate
from langchain.output_parsers import ResponseSchema
from langchain.output_parsers import StructuredOutputParser
```

```python
prompt_template = PromptTemplate(input_variables=['input', 'response'], template=task_template)
```

```python
message = prompt_template.format(input = train_df['text'][0], response= train_df['label'][0])
```

```python
pairs = []
for row in train_df.itertuples():
    text = prompt_template.format(input = row[1], response= row[2])
    pairs.append({"text":text})
```

```python
## Model training steps
```

```python
import torch
from transformers import AutoModelForCausalLM, BitsAndBytesConfig, AutoTokenizer, TrainingArguments
from peft import LoraConfig
from trl import SFTTrainer
```

```python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
)
```

```python
base_model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map=device,
    trust_remote_code=True,
    use_auth_token=True
)
```

```python
base_model.config.use_cache = False
base_model.config.pretraining_tp = 1
peft_config = LoraConfig(
    lora_alpha=16,
    lora_dropout=0.1,
    r=64,
    bias="none",
    task_type="CAUSAL_LM",
)
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token
```

```python
output_dir = "./results"
```

```python
training_args = TrainingArguments(
    output_dir=output_dir,
    per_device_train_batch_size=4,
    gradient_accumulation_steps=4,
    learning_rate=2e-4,
    logging_steps=10,
    max_steps=500
)
```

```python
# max_seq_length = 512
```

```python
data_files = {"train": "train.jsonl"}
dataset_name = "abipani/ag_news_jsonl"
dataset = load_dataset(dataset_name, data_files=data_files)
```

```python
dataset['train']
```

```python
trainer = SFTTrainer(
    model=base_model,
    train_dataset=dataset['train'],
    peft_config=peft_config,
    dataset_text_field="text",
    max_seq_length=256,
    tokenizer=tokenizer,
    args=training_args,
)
```
